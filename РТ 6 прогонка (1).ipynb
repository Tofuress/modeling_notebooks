```python
a = [[5, 3, 0, 0],
     [3, 6, 1, 0],
     [0, 1, 4, -2],
     [0, 0, 1, -3]]
A = [0, 3, 1, 1]
B = [3, 1, -2, 0]
C = [5, 6, 4, -3]
F = [8, 10, 3, -2]
```

```python
def func(a, b, c, f):
    # a - нижняя
    # b - верхняя
    # с - главная
    # f - правая
    a, b, c, f = tuple(map(lambda x: list(map(float, x)), (a, b, c, f)))

    alpha = [-b[0] / c[0]]
    beta = [f[0] / c[0]]
    n = len(f)
    x = [0] * n

    for i in range(1, n):
        # каждый шаг вычисляет коэффициенты для xi = alphai * xi+1 + betai
        # znamenatel = a1 * alpha0 + c1
        # alpha[1] = -b1 / znamenatel
        # beta[1] = (f1 - a1 * beta0) / znamenatel
        alpha.append(-b[i] / (a[i] * alpha[i - 1] + c[i]))
        beta.append((f[i] - a[i] * beta[i - 1]) / (a[i] * alpha[i - 1] + c[i]))
    print(alpha)
    print(beta)

    x[n - 1] = beta[n - 1]

    # обратный ход
    for i in range(n - 1, 0, -1):
        # x2 = alpha2 * x3 + beta2
        x[i - 1] = alpha[i - 1] * x[i] + beta[i - 1]
        
    print('Решение:')
    return x

print(func(A, B, C, F))
```
