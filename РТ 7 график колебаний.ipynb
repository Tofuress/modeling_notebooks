```python
import numpy as np
import matplotlib.pyplot as plt

T0 = 0.001
A = 1
omega = 0.2
phi0 = 0
delta = 0.01
N = 100000

n = np.arange(N)
t = T0 * n
w = np.sqrt(omega**2 - delta**2)
y = A * np.exp(-delta * t) * np.sin(w * t + phi0)

for i in range(10):
    print(f"t[{i}] = {t[i]:.5f}, y[{i}] = {y[i]:.6f}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(t, y, linewidth=0.8, color='b')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
