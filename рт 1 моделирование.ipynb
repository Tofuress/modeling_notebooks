```python
import numpy as np
import matplotlib.pyplot as plt
from math import gcd

R, r = 500, 3
p = R // gcd(R,r)

f = np.arange(0.0, 2 * np.pi * p, 0.01)
k = (R - r) / r
#x = (R - r) * np.cos(f) + r * np.cos(k * f)
#y = (R - r) * np.sin(f) - r * np.sin(k * f)
x = r * (R/r - 1) * (np.cos(f) + (np.cos(((R/r - 1) * f))) / (R/r - 1))
y = r * (R/r - 1) * (np.sin(f) - (np.sin(((R/r - 1) * f))) / (R/r - 1))

plt.figure(figsize=(6, 6))
plt.gca().set_aspect('equal','box')
plt.axis('equal')
plt.grid(True)
plt.plot(x, y)
#plt.xlim(-10, 10)
#plt.ylim(-10, 10)
```
